```python
df = pd.read_parquet("abfss://ML_Stab_TI_Energinet_EBI@onelake.dfs.fabric.microsoft.com/Ines.Lakehouse/Tables/mw08003_complete")
df.sort_values('Zeitpunkt', inplace= True)
df = df.dropna()
df['Wert'] = df['Wert'].astype(float)
df['temp'] = df['temp'].astype(float)
df.reset_index(drop=True, inplace=True)
```

```python
sequence_length = 24
numerical_cols = ["Wert", "temp"]
categorical_cols = [col for col in df.columns if col not in numerical_cols]
```

```python
# # Trim the DataFrame
df_trimmed = df.iloc[:-(df.shape[0] % sequence_length)] #  The number of samples must be a multiple of the sequence length
print('shape of trimed: ', df_trimmed.shape)
```

```python
# Assuming df is your DataFrame and the columns are 'col1' and 'col2'
scaler = MinMaxScaler()
df_trimmed[numerical_cols] = scaler.fit_transform(df_trimmed[numerical_cols])
```

```python
df_trimmed
```

```python
# Defining model and training parameters
model_args = ModelParameters(batch_size=100,
                             lr=0.001,
                             betas=(0.2, 0.9),
                             latent_dim=20,
                             gp_lambda=2,
                             pac=1)
 
train_args = TrainParameters(epochs=400,
                             sequence_length=sequence_length,
                             sample_length=8,
                             rounds=1,
                             measurement_cols=["Wert", "temp"])
```

```python
# Training the DoppelGANger synthesizer
model_dop_gan = TimeSeriesSynthesizer(modelname='doppelganger',model_parameters=model_args)
model_dop_gan.fit(df_trimmed, train_args, num_cols=numerical_cols, cat_cols=categorical_cols)
```

```python
# # Generate synthetic data
n = int(len(df_trimmed)/sequence_length)
print(n)
synth_data = model_dop_gan.sample(n_samples=n)
synth_df = pd.concat(synth_data, axis=0)
```

```python
# Create a plot for each measurement column
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(df_trimmed['Wert'].reset_index(drop=True), label='Real')
plt.plot(synth_df['Wert'].reset_index(drop=True), label='Synthetic', alpha=0.7)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Wert Comparison')
plt.legend()
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(df_trimmed['temp'].reset_index(drop=True), label='Real')
plt.plot(synth_df['temp'].reset_index(drop=True), label='Synthetic', alpha=0.7)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('temp Comparison')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Assuming df_trimmed and synth_df are predefined and normalized dataframes

# Initialize lists to hold the statistics
ks_results = []
pearson_correlations = []
tvd_results = []

# Compute KS, Pearson, and TVD for each attribute/column separately
for column in df_trimmed.columns:
    # KS Test
    ks_stat, ks_pvalue = ks_2samp(df_trimmed[column], synth_df[column])
    ks_results.append((column, ks_stat, ks_pvalue))
    
    # Pearson Correlation
    corr, _ = pearsonr(df_trimmed[column], synth_df[column])
    pearson_correlations.append((column, corr))
    
    # TVD Calculation
    p, _ = np.histogram(df_trimmed[column], bins=10, density=True)
    q, _ = np.histogram(synth_df[column], bins=10, density=True)
    tvd_value = np.sum(np.abs(p - q)) / 2
    tvd_results.append((column, tvd_value))

# Split PCA and t-SNE visualizations
pca = PCA(n_components=2)
df_trimmed_pca = pca.fit_transform(df_trimmed)
synth_df_pca = pca.transform(synth_df)

tsne = TSNE(n_components=2, random_state=42)
df_trimmed_tsne = tsne.fit_transform(df_trimmed)
synth_df_tsne = tsne.fit_transform(synth_df)

# Selecting indices to segment the data for PCA and t-SNE plots
# Replace these with the specific indices you're interested in
indices_segment = [0, len(df_trimmed) // 4, len(df_trimmed) // 2, 3 * len(df_trimmed) // 4, len(df_trimmed)]

fig, axs = plt.subplots(4, 2, figsize=(15, 20))

for i in range(4):
    idx_range = range(indices_segment[i], indices_segment[i+1])
    # PCA Plot for segment
    axs[i, 0].scatter(df_trimmed_pca[idx_range, 0], df_trimmed_pca[idx_range, 1], color='red', label='Original')
    axs[i, 0].scatter(synth_df_pca[idx_range, 0], synth_df_pca[idx_range, 1], color='blue', label='Synthetic')
    axs[i, 0].set_title(f'PCA Segment {i+1}')
    axs[i, 0].legend()

    # t-SNE Plot for segment
    axs[i, 1].scatter(df_trimmed_tsne[idx_range, 0], df_trimmed_tsne[idx_range, 1], color='red', label='Original')
    axs[i, 1].scatter(synth_df_tsne[idx_range, 0], synth_df_tsne[idx_range, 1], color='blue', label='Synthetic')
    axs[i, 1].set_title(f't-SNE Segment {i+1}')
    axs[i, 1].legend()

plt.tight_layout()
plt.show()

# Output the statistics
print('KS Test Results:')
print(*ks_results, sep='\n')

print('\nPearson Correlation Results:')
print(*pearson_correlations, sep='\n')

print('\nTVD Results:')
print(*tvd_results, sep='\n')

```

```python
def create_rolling_windows(df, window_size):
    """
    Create rolling windows of the specified size from the DataFrame.
    Each window will be a separate sequence in the resulting array.
    """
    rolling_windows = []
    for start in range(len(df) - window_size + 1):
        end = start + window_size
        rolling_windows.append(df[start:end].values)
    return np.array(rolling_windows)

# Apply rolling window transformation
window_size = 24
df_trimmed_rolling = create_rolling_windows(df_trimmed, window_size)
synth_df_rolling = create_rolling_windows(synth_df, window_size)
```

```python
def create_split_indices(total_samples, test_size, validation_size):
    """
    Create indices for the training, validation, and testing splits.
    """
    indices = np.arange(total_samples)
    np.random.shuffle(indices)  # Shuffle the indices
    test_end = int(total_samples * test_size)
    validation_end = test_end + int(total_samples * validation_size)
    
    test_indices = indices[:test_end]
    validation_indices = indices[test_end:validation_end]
    train_indices = indices[validation_end:]
    return train_indices, validation_indices, test_indices

def split_data_with_indices(processed_data, train_indices, validation_indices, test_indices, target_col_indices):
    """
    Split data into training, validation, and testing sets using provided indices and multiple target columns.
    """
    # Extract features from all timesteps except the last one for the training set
    X_train = processed_data[train_indices, :-1, :]
    
    # Targets are from the last timestep of each sequence for the specified target columns for the training set
    y_train = processed_data[train_indices][:, -1, :][:, target_col_indices]
    
    # Perform similar operations for validation and test sets
    X_validation = processed_data[validation_indices, :-1, :]
    y_validation = processed_data[validation_indices][:, -1, :][:, target_col_indices]
    
    X_test = processed_data[test_indices, :-1, :]
    y_test = processed_data[test_indices][:, -1, :][:, target_col_indices]

    return X_train, X_validation, X_test, y_train, y_validation, y_test


# Create split indices based on the total number of sequences
total_sequences = df_trimmed_rolling.shape[0]
train_indices, validation_indices, test_indices = create_split_indices(
    total_samples=total_sequences,
    test_size=0.10,
    validation_size=0.15
)

# Assuming target_col_indices = [0, 1] because 'wert' is at index 0 and 'temp' is at index 1
target_col_indices = [0,1]

# Split the data using the indices
X_real_train, X_real_validation, X_real_test, y_real_train, y_real_validation, y_real_test = split_data_with_indices(
    processed_data=df_trimmed_rolling,
    train_indices=train_indices,
    validation_indices=validation_indices,
    test_indices=test_indices,
    target_col_indices=target_col_indices
)

X_synth_train, X_synth_validation, X_synth_test, y_synth_train, y_synth_validation, y_synth_test = split_data_with_indices(
    processed_data=synth_df_rolling,
    train_indices=train_indices,
    validation_indices=validation_indices,
    test_indices=test_indices,
    target_col_indices=target_col_indices
)

```

```python
def CNN_GRU_regression(timesteps, features_per_timestep, units):
    model = Sequential()

    # CNN Layers
    # Assume the input shape is (timesteps, features_per_timestep)
    model.add(Conv1D(filters=4, kernel_size=2, activation='relu', input_shape=(timesteps, features_per_timestep)))
    model.add(MaxPooling1D(pool_size=2))

    model.add(Conv1D(filters=16, kernel_size=2, activation='relu'))
    model.add(MaxPooling1D(pool_size=2))

    # Reshape the output to be compatible with GRU layers
    # Calculate the new number of timesteps and features after pooling layers
    new_timesteps = timesteps // (2 * 2) # Assuming pool_size=2 for both pooling layers
    new_features = 16 # This is the number of filters in the last Conv1D layer
    model.add(Reshape((new_timesteps, new_features)))

    model.add(Dropout(0.5))

    # GRU Layer - no need for Flatten since we want to process temporal sequences
    model.add(GRU(units=units, return_sequences=False))

    # Output Layer
    model.add(Dense(units=units))  

    # Compile the model
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
    
    return model

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
# Initialize the model
timesteps = sequence_length - 1
units = 2
feature_n = 2

real_model = CNN_GRU_regression(timesteps, feature_n, units)

# Train the model on real

history = real_model.fit(
    X_real_train, y_real_train,
    validation_data=(X_real_validation, y_real_validation),
    epochs=100,  # Adjust the number of epochs as needed
    batch_size=128,  # Adjust the batch size as needed
    callbacks=[early_stopping]
)
```
